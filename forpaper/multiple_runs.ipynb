```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from plotting_utils import TI_per_pair_plot_with_confidence_intervals
```

```python
pred_mats_icl = np.load("pred_mats_10runs_icl.npy", allow_pickle=True)
pred_mats_iwl = np.load("pred_mats_10runs_iwl.npy", allow_pickle=True)
pred_mats_icl_linear = np.load("pred_mats_10runs_icl_linear.npy", allow_pickle=True)
```

```python
sns.set_context("paper", font_scale=1.9, rc={"lines.linewidth": 3})

fig, ax = plt.subplots(3, 1, sharex=True, figsize=(10, 8))

fig.supylabel('Prediction value')
fig.supxlabel('Query pair')

for i, pred_mats in enumerate([pred_mats_iwl, pred_mats_icl, pred_mats_icl_linear]):
    TI_per_pair_plot_with_confidence_intervals(pred_mats, ax=ax[i])
    if i == 0:
        ax[i].set_title("IWL")
    elif i == 1:
        ax[i].set_title("ICL")
    else:
        ax[i].set_title("ICL linear")
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')

plt.tight_layout()

plt.savefig("multiple_runs.pdf", dpi=300, bbox_inches='tight')
```
